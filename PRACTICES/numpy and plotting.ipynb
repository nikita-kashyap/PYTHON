```python
#        Box_Plot

import matplotlib.pyplot as plt

value1=[84,77,20,67,62,75,78,71,32,38,89,78,67,72,82,87,66,56,52,5]
value2=[62,5,91,25,35,32,96,99,3,90,95,34,27,55,100,67,35,45,78,8]
value3=[67,45,68,98,24,56,12,78,95,37,81,40,55,6,7,87,35,67,62,22]
value4=[65,77,87,23,45,98,23,67,98,12,3,87,56,64,65,99,100,77,88]

box_plot_data=[value1,value2,value3,value4]
plt.boxplot(box_plot_data)
plt.show()
```

```python
value1=[84,77,20,67,62,75,78,71,32,38,89,78,67,72,82,87,66,56,52,5]
value2=[62,5,91,25,35,32,96,99,3,90,95,34,27,55,100,67,35,45,78,8]
value3=[67,45,68,98,24,56,12,78,95,37,81,40,55,6,7,87,35,67,62,22]
value4=[65,77,87,23,45,98,23,67,98,12,3,87,56,64,65,99,100,77,88]

box_plot_data=[value1,value2,value3,value4]
plt.boxplot(box_plot_data,patch_artist=True,labels=["subject1","subject2","subject3","subject4"])
plt.show()
# lines indicate mean value
```
